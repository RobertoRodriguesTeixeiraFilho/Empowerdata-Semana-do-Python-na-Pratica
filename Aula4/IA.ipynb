# Inteligência Artificial (IA)

```python
import yfinance as yf
```

## Coleta de dados

```python
dados = yf.Ticker("PETR4.SA").history("2y")
```

## Tratamento de dados

```python
treinamento = dados.reset_index()
treinamento["Date"] = treinamento["Date"].dt.date
treinamento = treinamento[["Date", "Close"]]
# Nomeando as colunas: ds é a Abreviação de Dates;
treinamento.columns = ["ds", "y"]
treinamento
```

## Criação do modelo de Machine Learning (Aprendizado de Máquina)

```python
from prophet import Prophet
```

```python
modelo = Prophet()
```

## Treinando o modelo de ML (Machine Learning)

```python
modelo.fit(treinamento)
```

## Gerando datas futuras

```python
datas_futuras = modelo.make_future_dataframe(30)
```

## Gerando as previsões

```python
prev = modelo.predict(datas_futuras)
prev
```

## Visualização das previsões

```python
from prophet.plot import plot_plotly
```

```python
grafico_prev = plot_plotly(modelo, prev)
grafico_prev.show()
```

### Salvando a previsão

```python
grafico_prev.write_html("Grafico-previsto.html")
```
